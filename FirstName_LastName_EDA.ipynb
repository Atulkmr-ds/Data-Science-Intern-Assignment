```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("Customers.csv")
print("Dataset Info:")
print(data.info())

# Step 2: Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Step 3: Convert SignupDate to datetime
data['SignupDate'] = pd.to_datetime(data['SignupDate'])

# Step 4: Summary statistics
print("\nSummary Statistics:")
print(data.describe(include="all"))

# Step 5: Analyze unique values in 'Region'
print("\nUnique Values in 'Region':")
region_counts = data['Region'].value_counts()
print(region_counts)

# Step 6: Visualizations

# Distribution of regions
plt.figure(figsize=(8, 5))
sns.barplot(x=region_counts.index, y=region_counts.values, palette="viridis")
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()

# Signup trends over time
signups_over_time = data['SignupDate'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
signups_over_time.plot(kind="line", marker="o", color="blue")
plt.title("Signup Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Signups")
plt.grid()
plt.show()

# Boxplot of signups by region
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x="Region", y=data["SignupDate"].dt.year, palette="coolwarm")
plt.title("Signups by Region Over Years")
plt.xlabel("Region")
plt.ylabel("Signup Year")
plt.xticks(rotation=45)
plt.show()

# Step 7: Save cleaned and processed dataset
data.to_csv("Processed_Customers.csv", index=False)
print("\nProcessed dataset saved as 'Processed_Customers.csv'")
```
